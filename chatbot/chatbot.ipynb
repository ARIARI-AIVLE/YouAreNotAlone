```python
#전처리 1-1
import re
def del_newline(text : str):
    return re.sub('[\s\n\t]+', ' ', text)

def del_special_char(text : str):
    return re.sub('[^가-힣ㄱ-ㅎㅏ-ㅣ.0-9a-zA-Z\s?!:/]+', ' ', text)

def del_duplicated_space(text : str):
    return re.sub('[\s]+', ' ', text)

def preprocess(text : str):
    proc_txt = del_newline(text)
    proc_txt = del_special_char(proc_txt)
    proc_txt = del_duplicated_space(proc_txt)
    return proc_txt.strip()
```

```python
교육부_1 = 교육부_1.apply(lambda x: x.apply(lambda y: preprocess(y)))
교육부_2 = 교육부_2.apply(lambda x: x.apply(lambda y: preprocess(y)))
지원센터 = 지원센터.apply(lambda x: x.apply(lambda y: preprocess(y)))
폭력예방재단 = 폭력예방재단.apply(lambda x: x.apply(lambda y: preprocess(y)))
상담복지개발원 = 상담복지개발원.apply(lambda x: x.apply(lambda y: preprocess(y)))
```

```python
#전처리 1-2 : 띄어쓰기와 맞춤법
def spell_check(x):
    if len(x)>500:
        return spell_checker.check(x[:500]).checked + spell_check(x[500:])
    else:
        return spell_checker.check(x).checked
```

```python
col = ['제목', '질문']

for j in col:
    교육부_1[j] = 교육부_1[j].apply(lambda x: spell_check(spacing(x)))
    교육부_2[j] = 교육부_2[j].apply(lambda x: spell_check(spacing(x)))
    지원센터[j] = 지원센터[j].apply(lambda x: spell_check(spacing(x)))
    폭력예방재단[j] = 폭력예방재단[j].apply(lambda x: spell_check(spacing(x)))
    상담복지개발원[j] = 상담복지개발원[j].apply(lambda x: spell_check(spacing(x)))
```

```python
#전처리 1-3 : title 과 question을 merge
def t_q_merge(data)
    data['제목'] = data['제목'].fillna('')
    data['질문']  = data['질문'].fillna('')
    data['Q'] = data['제목'] + '\n' + data['질문']
    data.drop(['제목','질문'], axis=1, inplace=True)
    data.columns = ['A', 'Q']
    return data
```

```python
import requests
import json

def busan_correct(text):
    text = text.replace('\n',' ')
    response = requests.post('http://164.125.7.61/speller/results', data={'text1': text})
    data = response.text.split('data = [', 1)[-1].rsplit('];', 1)[0]
    if '맞춤법과 문법 오류를 찾지' in data or '결과를 받지 못했습니다.' in data:
        return text
    data = json.loads(data)
    
    flag = 0

    for err in data['errInfo']:
        start=err['start']
        end=err['end']
        before = text[start+flag : end+flag]
        if before.isalpha(): #영어는 교정하지 않음
            continue
        after = err['candWord'].split('|')[0]
        
        #print(f'{before}->{after}_')
        
        text = text[:start+flag] + after + text[flag+end:]
        
        if len(before) < len(after): #바뀐 후의 자릿수가 더 많으면 그만큼 더해줘야함
            flag += (len(after) - len(before))
        else: #바뀌기 전의 자릿수가 더 많으면 flag를 빼줘야함
            flag -= (len(before) - len(after))
        
    return text
```

```python
import numpy as np
import pandas as pd
import torch
from transformers.optimization import AdamW, get_cosine_schedule_with_warmup
from transformers import PreTrainedTokenizerFast, GPT2LMHeadModel
from torch.utils.data import DataLoader, Dataset
import re
```

```python
Chatbot_Data = pd.read_csv('/content/drive/MyDrive/기타/청소년상담복지 공모전/DATA/Chatbot_Data.csv', encoding='CP949')
```

```python
Chatbot_Data2 = Chatbot_Data2[~Chatbot_Data2['A'].str.endswith('군요.')]
Chatbot_Data = pd.concat([Chatbot_Data, Chatbot_Data2])
```

```python
Q_TKN = "<usr>"
A_TKN = "<sys>"
BOS = '</s>'
EOS = '</s>'
MASK = '<unused0>'
SENT = '<unused1>'
PAD = '<pad>'
```

```python
koGPT2_TOKENIZER = PreTrainedTokenizerFast.from_pretrained("skt/kogpt2-base-v2",
            bos_token=BOS, eos_token=EOS, unk_token='<unk>',
            pad_token=PAD, mask_token=MASK) 
model = GPT2LMHeadModel.from_pretrained('skt/kogpt2-base-v2')
```

```python
class ChatbotDataset(Dataset):
    def __init__(self, chats, max_len=40):  
        self._data = chats
        self.max_len = max_len
        self.q_token = Q_TKN
        self.a_token = A_TKN
        self.sent_token = SENT
        self.eos = EOS
        self.mask = MASK
        self.tokenizer = koGPT2_TOKENIZER

    def __len__(self):  
        return len(self._data)

    def __getitem__(self, idx):  
        turn = self._data.iloc[idx]
        q = turn["Q"]  
        a = turn["A"]  
        
        q_toked = self.tokenizer.tokenize(self.q_token + q + self.sent_token)
        q_len = len(q_toked)

        a_toked = self.tokenizer.tokenize(self.a_token + a + self.eos)
        a_len = len(a_toked)

        #질문의 길이가 최대길이보다 크면
        if q_len > self.max_len:
            a_len = self.max_len - q_len        
            if a_len <= 0:       
                q_toked = q_toked[-(int(self.max_len / 2)) :]   
                q_len = len(q_toked)
                a_len = self.max_len - q_len      
            a_toked = a_toked[:a_len]
            a_len = len(a_toked)

        #질문의 길이 + 답변의 길이가 최대길이보다 크면
        if q_len + a_len > self.max_len:
            a_len = self.max_len - q_len      
            if a_len <= 0:    
                q_toked = q_toked[-(int(self.max_len / 2)) :]  
                q_len = len(q_toked)
                a_len = self.max_len - q_len            
            a_toked = a_toked[:a_len]
            a_len = len(a_toked)

        # 답변 labels = [mask, mask, ...., mask, ..., <bos>,..답변.. <eos>, <pad>....]
        labels = [self.mask,] * q_len + a_toked[1:]

        # mask = 질문길이 0 + 답변길이 1 + 나머지 0
        mask = [0] * q_len + [1] * a_len + [0] * (self.max_len - q_len - a_len)
        # 답변 labels을 index 로 만든다.
        labels_ids = self.tokenizer.convert_tokens_to_ids(labels)
        # 최대길이만큼 PADDING
        while len(labels_ids) < self.max_len:
            labels_ids += [self.tokenizer.pad_token_id]

        # 질문 + 답변을 index 로 만든다.    
        token_ids = self.tokenizer.convert_tokens_to_ids(q_toked + a_toked)
        # 최대길이만큼 PADDING
        while len(token_ids) < self.max_len:
            token_ids += [self.tokenizer.pad_token_id]

        #질문+답변, 마스크, 답변
        return (token_ids, np.array(mask), labels_ids)
```

```python
def collate_batch(batch):
    data = [item[0] for item in batch]
    mask = [item[1] for item in batch]
    label = [item[2] for item in batch]
    return torch.LongTensor(data), torch.LongTensor(mask), torch.LongTensor(label)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
train_set = ChatbotDataset(Chatbot_Data, max_len=100)
train_dataloader = DataLoader(train_set, batch_size=8, num_workers=0, shuffle=True, collate_fn=collate_batch,)
```

```python
model.to(device)
model.train()
```

```python
learning_rate = 3e-5
criterion = torch.nn.CrossEntropyLoss(reduction="none")
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

epoch = 8
Sneg = -1e18
```

```python
import time
```

```python
print ("start")
for epoch in range(epoch):
    start_time = time.time()

    for batch_idx, samples in enumerate(train_dataloader):
        optimizer.zero_grad()
        token_ids, mask, label = samples
        
        if device.type=='cuda':
            token_ids = token_ids.to('cuda:0')
            mask = mask.to('cuda:0')
            label = label.to('cuda:0')

        out = model(token_ids)
        out = out.logits     
        mask_3d = mask.unsqueeze(dim=2).repeat_interleave(repeats=out.shape[2], dim=2)
        mask_out = torch.where(mask_3d == 1, out, Sneg * torch.ones_like(out))
        loss = criterion(mask_out.transpose(2, 1), label)
        # 평균 loss 만들기 avg_loss[0] / avg_loss[1] <- loss 정규화
        avg_loss = loss.sum() / mask.sum()
        avg_loss.backward()
        # 학습 끝
        optimizer.step()

    torch.save(model.state_dict(), '/content/drive/MyDrive/기타/청소년상담복지 공모전/Chatbot/chatbot_지식인감성대화_8.pt')
    end_time = time.time()
    print(f'epoch {epoch} end, time : {end_time - start_time}')
print ("end")
```
