```python
result = pd.DataFrame(df.groupby(['시도', '행정구']).size()).reset_index()
result.columns = ['시도','행정구', '응답']

#사이버 폭력 경험이 있는 학생을 시도별로 count (9라고 답변한 학생 제외)
for i in range(1,9):
    temp = df[df[f'Q8a_{i}'] != 9].groupby(['시도', '행정구'])[f'Q8a_{i}'].count().reset_index()
    temp.columns = ['시도','행정구', f'Q8a_{i}']

    result = pd.merge(result, temp, on=['시도','행정구'])
    result[f'Q8a_{i}'] /= result['응답'] / 100 #percentage
result
```

### (2) 행정구 별


#### 행정구별 질문별 percentage

```python
result = pd.DataFrame(df.groupby(['시도', '행정구']).size()).reset_index()
result.columns = ['시도','행정구', '응답']

#사이버 폭력 경험이 있는 학생을 시도별로 count (9라고 답변한 학생 제외)
for i in range(1,9):
    temp = df[df[f'Q9a_{i}'] != 9].groupby(['시도', '행정구'])[f'Q9a_{i}'].count().reset_index()
    temp.columns = ['시도','행정구', f'Q9a_{i}']

    result = pd.merge(result, temp, on=['시도','행정구'])
    result[f'Q9a_{i}'] /= result['응답'] / 100 #percentage
result
```

#### 피해 경험이 많은 행정구

```python
result['시도_행정구'] = result['시도'] + ' ' +result['행정구']
```

```python
# 질문 별로 피해 경험이 많은 시도 나열
top_sido = result.sort_values(by='Q9a_1', ascending=False)['시도_행정구'].reset_index(drop=True)
for i in range(2,9):
    temp = result.sort_values(by=f'Q9a_{i}', ascending=False)['시도_행정구'].reset_index(drop=True)
    top_sido = pd.concat([top_sido, temp], axis=1).head(10)
top_sido.columns = [f'Q9a_{i}' for i in range(1,9)]
top_sido
```

대부분 조사인원 너무 적어서 상위권으로 오름,,, 시도 별로 보는 게 좋을 듯

#### 시도별 피해 경험 많은 행정구

```python
def region(name:str):
    region = result[result['시도'] == name]

    top_region = region.sort_values(by='Q9a_1', ascending=False)['행정구'].reset_index(drop=True)
    for i in range(2,9):
        temp = region.sort_values(by=f'Q9a_{i}', ascending=False)['행정구'].reset_index(drop=True)
        top_region = pd.concat([top_region, temp], axis=1).head(5)
    top_region.columns = [f'Q9a_{i}' for i in range(1,9)]
    return top_region
```
